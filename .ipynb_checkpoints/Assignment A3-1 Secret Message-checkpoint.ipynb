## Defining Functions

### `read_filenames`
Reads the filenames from a given path and returning a list of filenames appended to the path as a relative path.

_Parameters_  
`path: str` the relative path to the directory.

_Returns_ `list` with relative filepaths.

````python
def read_filenames(path: str) -> list:
    filenames = os.listdir(path)
    return [path + filename for filename in filenames]
````

### `create_kvpairs`
Takes a list and creates a key value pair with the filename as key and the relative path as value.  
This functions uses a Regular expression to isolate the relevant value from the filepath.

_[reference, Regular expression operations Python](https://docs.python.org/2/library/re.html)_

_Parameters_  
`filepaths: list` with relative filepaths.

_Returns_ `dict` with formatted filename as key and relative filepaths as value.

````python
def create_kvpairs(filepaths: list) -> dict:
    files = dict()
    for filepath in filepaths:
        filename = re.sub('.+[\d]', '', filepath)
        files[filename] = filepath   
        
    return files
````

### `plot_files`
Takes the dictionary with name and path and iterates through a `sorted` for loop.  
`cv2` is used to read the files, and `pyplot` to plot them.

_Parameter_  
`files: dict` filename as key and relative filepaths as value.

````python
def plot_files(files: dict):
    for file_key in sorted (files):
        file = cv2.imread(files[file_key]) 
        plt.imshow(file)
        plt.show()
````

### `main`
Takes the relative path to the directory and calls `read_filenames()`, `create_kvpairs` and `plot_files`.

_Parameters_  
`path: str` relative path to directory.

````python
def main(path: str):    
    filenames = read_filenames(path)
    file_kv = create_kvpairs(filenames)
    plot_files(file_kv)
    capture_image()
````

## Utility Function
### `capture_image`
Opens your webcam with `cv2` and captures a picture and then shows it with `pyplot`.

````python
def capture_image():    
    cap = cv2.VideoCapture(0)
    ret, image = cap.read()
    cap.release()
    print('You hadn\'t seen that coming :)')
    plt.imshow(image)
    plt.show()
````
